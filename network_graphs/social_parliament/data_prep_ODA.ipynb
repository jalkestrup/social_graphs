```python
#import json file sample_json.json

import json
with open('sample_json.json') as f:
    data_sample = json.load(f)


import pandas as pd
import numpy as np


def voting_df(data):
    df = pd.DataFrame()

    df['afstemnings_id'] = [x['id'] for x in data['value']]
    df['sagstrin_id'] = [x['sagstrinid'] for x in data['value']]
    df['vedtaget'] = [x['vedtaget'] for x in data['value']]
    df['dato'] = [x['opdateringsdato'] for x in data['value']]
    df['sag_id'] = [x['Sagstrin']['Sag']['id'] for x in data['value']]
    df['titel']= [x['Sagstrin']['Sag']['titel'] for x in data['value']]
    df['titel_kort'] = [x['Sagstrin']['Sag']['titelkort'] for x in data['value']]
    df['lovnummer_num'] = [x['Sagstrin']['Sag']['nummernumerisk'] for x in data['value']]
    df['lovnummer'] = [x['Sagstrin']['Sag']['nummer'] for x in data['value']]
    df['resume']= [x['Sagstrin']['Sag']['resume'] for x in data['value']]
    df['periode_id']= [x['Sagstrin']['Sag']['periodeid'] for x in data['value']]

    return df
    

        
df = voting_df(data_sample)
df.head()
```
